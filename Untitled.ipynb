```python
# Monte Carlo Simulation

import random
import matplotlib.pyplot as plt

n_samples = 10000
n_circle = 0
plt.figure()

for i in range(n_samples):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    
    if x**2 + y**2 <= 1:
        n_circle = n_circle + 1
        plt.plot(x,y, 'o', color='blue', markersize=1)
    else:
        plt.plot(x,y, 'o', color='red', markersize=1)

plt.axis('equal')     
plt.show  

pi_estimate = 4 * n_circle / n_samples
print(f'pi_estimate = {pi_estimate}')

# The area of a circle is defined as pi*radius^2. That means pi = circle area / r^2. Notice that r^2 is the area of a square of side r (let's call it the small square. hence,the last formula can now be written as: pi = circle area / small square area. Notice that the area of the small square is 1/4  the area of the big one. Hence we can write pi = 4 * circle area / big square area. On the other hand, if we randomly generate points inside the big square, the number of points inside the circle is supposed to be proportional to the area of the circle. Similarly, the total number of points within the big square is supposed to be proportional to the area of the square. Hence pi = 4 * circle area / big square area = 4 * n_inside_circle / n_total   )
```
